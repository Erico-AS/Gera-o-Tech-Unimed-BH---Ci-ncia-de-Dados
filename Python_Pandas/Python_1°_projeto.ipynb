```python
#Menor item da lista
min(lista)
```

```python
#ordenar a lista
lista.extend([100,90])
lista
lista.sort()
lista
```

# ***Dicionários***

```python
#ultilizamos {} para criar
dc = {"Banana":15, "Maçã":10, "Uva":12} #dentro das {} vem o condeito, chave, valor
```

```python
#Acessar valor através de chave
dc["Banana"]
```

```python
#Vendo todas as chaves
dc.keys()
```

```python
#Mudando valor de uma chave
dc["Uva"] = 8
dc
```

```python
#Todos os valores do dicionário
dc.values()
```

```python
#Verificar se já existe uma certa 'chave' e 'valor' e se não existir ele irá adicionar
dc.setdefault("Laranja", 20)
dc
```
